```python
from glob import glob
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sys
sys.path.append('../')
from uvot import filter_colors
from astropy.stats import sigma_clip
from source_names_dict import source_names_readable
```

```python
import matplotlib
matplotlib.rcParams['mathtext.fontset'] = 'stix'
matplotlib.rcParams['font.family'] = 'STIXGeneral'
```

```python
filter_colors['B'] = 'blue'
filter_colors['V'] = '#009C15'
filter_colors['U'] = '#622DA1'
```

```python
files = glob('../lightcurves/joined_all/*')
for fn in files:
    print(fn)
    simbad_name = fn.split('/')[-1][:-4]
    readable_name = source_names_readable[simbad_name]
    
    df = pd.read_csv(fn, dtype={'OBSID':str})
    
    fig, axes = plt.subplots(2,1,figsize=(15,10), sharex=True, facecolor='white')
    ax = axes[0]
    ax2 = axes[1]
    sub = df.copy()
    #ma = sigma_clip(df['FULL'], sigma=4, maxiters=3)
    #sub = df[~ma.mask] 

    ax.errorbar(sub.index, sub['FULL'], yerr=sub['FULL_ERR'], color='black', lw=1.0, ls='none', label='FULL', marker='.', markersize=0.75)
    ax.errorbar(sub.index, sub['SOFT'], yerr=sub['SOFT_ERR'], color='red', lw=1.0, ls='none', label='SOFT', marker='.', markersize=0.75)
    ax.errorbar(sub.index, sub['HARD'], yerr=sub['HARD_ERR'], color='blue', lw=1.0, ls='none', label='HARD', marker='.', markersize=0.75)
    
    df_ul = sub[sub['FULL_UL'].fillna(False)]
    ax.errorbar(df_ul.index, df_ul['FULL'], yerr=(df_ul['FULL'], [0]*len(df_ul)), color='cyan', lw=1.0, ls='none', label='UL', marker='v', markersize=2)
    
    
    ax_op = ax2
    ax_uv = ax2.twinx()
    ax_op.set_ylabel('UVOT Optical band rate (ct/s)')
    ax_uv.set_ylabel('UVOT UV band rate (ct/s)')
    
    
    
    
    for f in ['V', 'B', 'U', 'UVW1','UVM2','UVW2']:
        if f in ['UVW1','UVM2','UVW2']:
            a = ax_uv
        else:
            a = ax_op

        sub = df.copy()
        ma  = sigma_clip(sub[f'{f}'], sigma=2, maxiters=5)
        if ma.mask is False:
            sub = sub[~ma.mask]
        
        a.errorbar(sub.index, sub[f'{f}'], yerr=sub[f'{f}_ERR'], color=filter_colors[f], lw=1.0, ls='none', label=f'{f}', marker='.', markersize=3)
        

        
        """
        if f in ['V','U','B']:
            a.yaxis.set_label_position('left')
            a.yaxis.set_ticks_position('left')
            a.spines["left"].set_visible(True)
        elif f in ['UVW1','UVM2','UVW2']:
            a.yaxis.set_label_position('right')
            a.yaxis.set_ticks_position('right')
            a.spines["left"].set_visible(True)
        """
        
    ax.legend(loc='upper left')
    ax.set_ylim(ymin=0)
    ax.set_ylabel('XRT Count Rate (ct/s)')
    
    ax2.set_xlabel('# Datapoint')
    
    ax_op.set_ylim(ymin=0)
    ax_uv.set_ylim(ymin=0)
    
    ax_op.legend(loc='upper left')
    ax_uv.legend(loc='upper right')
    
    ax.set_xlim(xmin=0, xmax=len(df))
    ax2.set_xlim(xmin=0, xmax=len(df))
    ax.grid(axis='x')
    ax2.grid(axis='x')
    plt.subplots_adjust(hspace=0)
    ax.set_title(readable_name, size=20)
    plt.savefig(f'../figures/lightcurves/joined_all_seq_no_xrt_clip/{simbad_name}.pdf', bbox_inches='tight')
    plt.savefig(f'../figures/lightcurves/joined_all_seq_no_xrt_clip/{simbad_name}.png', bbox_inches='tight')
    plt.show()
```
