```python
from glob import glob
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from astropy.table import Table, unique
matplotlib.rcParams['mathtext.fontset'] = 'stix'
matplotlib.rcParams['font.family'] = 'STIXGeneral'
```

```python
fits_files = glob('../lightcurves/uvot/*.fits')
fits_files
```

```python
for fits in fits_files:
    
    simbad_name = fits.split('/')[-1].split(',')[0]
    uvot_filter = fits.split('/')[-1].split(',')[1].split('_')[-1][:-5]
    print(fits)
    print(simbad_name)
    print(uvot_filter)
    
    tab = Table.read(fits)
    tab = unique(tab, keys=['OBSID'], keep='first')
    
    tab['idx'] = range(len(tab))
    
    fig, ax = plt.subplots(figsize=(20, 5), facecolor='white')
    
    
    ax.errorbar(tab['idx'], tab['COI_SRC_RATE'], yerr=tab['COI_SRC_RATE_ERR'], color='black', lw=1.0, capsize=1.0, ls='none', marker='.', markersize=5, label='Good')
    
    
    #for i in range(len(df)):
    #    ax.text(i, -0.03, df['obsID'].iloc[i], rotation=90, size=3)
    
    ax.set_xticks(tab['idx'])
    ax.set_xticklabels(tab['OBSID'], rotation=90, size=2)
    ax.legend(loc='upper right')
    ax.set_xlim(-1,len(tab)+1)
    ax.set_ylabel(f'UVOT {uvot_filter} Rate (ct/s)')
    plt.savefig(f'../lightcurves/uvot_plot/{simbad_name},{uvot_filter},sequential.pdf', bbox_inches='tight')
    plt.savefig(f'../lightcurves/uvot_plot/{simbad_name},{uvot_filter},sequential.png', bbox_inches='tight')
    plt.show()
    print(fits)
    print(tab)

    
```
