```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def R_es(mdot):
    R_es = 5e3 * mdot**(2/3)
    return R_es

def R_pr(mdot, m1):
    R_pr = 1.9e2*mdot**(16/21)*m1**(-8/21)
    return R_pr
```

```python
_mdot = np.linspace(0,10,100)
_m1 = 1
```

```python
plt.figure(figsize=(6,6))

for _m1 in [1,2,5,10]:
    _R_es = R_es(_mdot)
    _R_pr = R_pr(_mdot, _m1)
    plt.plot(_mdot, _R_es)
    plt.plot(_mdot, _R_pr)

plt.xlabel(r'$\dot{m}$')
plt.ylabel(r'$R_{g}$')
```
