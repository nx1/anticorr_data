```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('../tables/xrt_centroids.csv', dtype={'obsid': str})
df
```

```python
df.groupby(['simbad_name']).count()
```

```python
df = df.drop_duplicates(subset='obsid', keep='first')
df
```

```python
df.groupby(['simbad_name'])['ra','dec'].agg(['mean', 'std'])
```

```python
for simbad_name in df['simbad_name'].unique():
    sub = df[df['simbad_name'] == simbad_name]
    
    fig, ax = plt.subplots(1,3, figsize=(15,5))
    print(sub)
    ax[1].set_title(f'{simbad_name} | N={len(sub)}')
    
    ax[0].hist(sub['ra'], bins=20)
    ax[0].set_ylabel('RA')
    ax[1].hist(sub['dec'], bins=20)
    ax[1].set_ylabel('DEC')
    ax[2].scatter(sub['ra'], sub['dec'], marker='x')
    ax[2].set_xlabel('RA')
    ax[2].set_ylabel('DEC')
    
    plt.show()
```
