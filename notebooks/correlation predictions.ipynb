```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import matplotlib
matplotlib.rcParams['mathtext.fontset'] = 'stix'
matplotlib.rcParams['font.family'] = 'STIXGeneral'
from matplotlib import patches
```

```python
def arc_patch(center, radius, theta1, theta2, resolution=50, **kwargs):
    # generate the points
    theta = np.linspace(np.radians(theta1), np.radians(theta2), resolution)
    points = np.vstack((radius*np.cos(theta) + center[0], 
                        radius*np.sin(theta) + center[1]))
    # build the polygon and add it to the axes
    poly = patches.Polygon(points.T, closed=True, **kwargs)
    return poly
```

```python
fig, ax = plt.subplots(figsize=(6,6), facecolor='white')
ax.arrow(0.1, 0.95, 0,-0.5, shape='full', lw=1, length_includes_head=True, head_width=0.05, facecolor='black')
ax.arrow(0.7, 0.8, -0.5,-0.5, shape='full', lw=1, length_includes_head=True, head_width=0.05, facecolor='black')
ax.arrow(0.6, 0.5, +0.35,-0.35, shape='full', lw=1, length_includes_head=True, head_width=0.05, facecolor='black')

plt.text(0.15,0.82, s='High i', size=20)

e1 = patches.Ellipse((0.25, 0.75), 0.2, 0.02, angle=0, linewidth=2, fill=True, color='darkviolet')
e5 = patches.Ellipse((0.25, 0.75), 0.1, 0.1, angle=0, linewidth=2, fill=True, color='darkgreen')
ax.add_patch(e5)
ax.add_patch(e1)


plt.text(0.4,0.4, s='Mid i', size=20)
e6 = patches.Ellipse((0.7, 0.6), 0.2, 0.1, angle=0, linewidth=2, fill=True, color='darkviolet')
e7 = patches.Ellipse((0.7, 0.6), 0.05, 0.05, angle=0, linewidth=2, fill=True, color='darkgreen')
e8 = patches.Ellipse((0.7, 0.6325), 0.025, 0.01, angle=0, linewidth=2, fill=True, color='cyan')
ax.add_patch(e6)
ax.add_patch(e8)
ax.add_patch(e7)



plt.text(0.6,0.3, s='Low i', size=20)

e2 = patches.Ellipse((0.7, 0.2), 0.1, 0.1, angle=0, linewidth=2, fill=True, color='darkgreen', label='Wind')
e3 = patches.Ellipse((0.7, 0.2), 0.02, 0.02, angle=0, linewidth=2, fill=True, color='cyan', label='Funnel')
e4 = patches.Ellipse((0.7, 0.2), 0.15, 0.15, angle=0, linewidth=2, fill=True, color='darkviolet', label='Disc')
ax.add_patch(e4)
ax.add_patch(e2)
ax.add_patch(e3)

ax.set_xlim(0,1)
ax.set_ylim(0,1)
ax.set_xticks([])
ax.set_yticks([])
ax.set_xlabel('X-Rays', size=20)
ax.set_ylabel('UV', size=20)
ax.legend()
plt.savefig('../figures/corr_mdot.pdf', bbox_inches='tight')
plt.savefig('../figures/corr_mdot.png', bbox_inches='tight')
plt.show()
```
