```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
```

```python
matplotlib.rcParams['mathtext.fontset'] = 'stix'
matplotlib.rcParams['font.family'] = 'STIXGeneral'
```

```python
df_0 = pd.read_csv('../model/transit_changing_inclination/i_0.csv')
df_22_5 = pd.read_csv('../model/transit_changing_inclination/i_22_5.csv')
df_45 = pd.read_csv('../model/transit_changing_inclination/i_45.csv')

df_0 = df_0[~(df_0['phase_ang'] == -20)]
df_22_5 = df_22_5[~(df_22_5['phase_ang'] == -20)]
df_45 = df_45[~(df_45['phase_ang'] == -20)]

df_0 = df_0[~(df_0['phase_ang'] == 20)]
df_22_5 = df_22_5[~(df_22_5['phase_ang'] == 20)]
df_45 = df_45[~(df_45['phase_ang'] == 20)]

df_0 = df_0[~(df_0['phase_ang'] == 0)]
df_22_5 = df_22_5[~(df_22_5['phase_ang'] == 0)]
df_45 = df_45[~(df_45['phase_ang'] == 0)]

df_45 = df_45[df_45['phase_ang'] < 25]
```

```python
plt.figure(figsize=(4,4), facecolor='white')
plt.plot(df_0['phase_ang'], df_0['area_ratio_irr_star'], label=r'i = $0^{\circ}$', ls='-', color='black')
plt.plot(df_22_5['phase_ang'], df_22_5['area_ratio_irr_star'], label=r'i = $22.5^{\circ}$', ls='--', color='black')
plt.plot(df_45['phase_ang'], df_45['area_ratio_irr_star'], label=r'i = $45^{\circ}$', ls='dotted', color='black')
#plt.plot(df_25['phase_ang'], df_25['area_ratio_irr_star'], label=r'i = $25^{\circ}$')

#plt.axhline(0)
#plt.ylim(0,0.07)
plt.ylim(0,0.15)
plt.xlabel('Orbital Phase ($^\circ$)')
plt.ylabel('Irradiated Fraction')
plt.legend()
plt.savefig('../figures/irr_area_inclination.pdf', bbox_inches='tight')
plt.savefig('../figures/irr_area_inclination.png', bbox_inches='tight')
```
