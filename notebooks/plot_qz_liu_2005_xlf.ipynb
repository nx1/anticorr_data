```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def calc_cum_hist(data, bins):
    """Calculate cumulative histogram.
    Sometimes known as the survival function."""
    hist = np.histogram(data, bins=bins)[0]
    hist_cum = np.cumsum(hist)
    hist_surv = sum(hist) - hist_cum
    # assert(hist.max() == N_sys) # Force full sample in hist
    return hist_surv
```

```python
df = pd.read_csv('../external/QZ_Liu_2005/table1b.csv')
df['lum'] = df['LX'] * 1e39
df
```

```python
plt.figure(figsize=(3.5,3.5))
plt.yscale('log')
```

```python
bins = np.arange(39,41,0.1)
bin_centers = 0.5 * (bins[:-1] + bins[1:])
Lx = np.log10(df['lum'])
Lx_cum = calc_cum_hist(Lx, bins)
```

```python
fig, ax = plt.subplots(2,2, figsize=(6,6))
ax[0][0].hist(Lx, bins=bins)
ax[0][1].hist(Lx, bins=bins)
ax[1][0].step(bin_centers, Lx_cum)
ax[1][1].step(bin_centers, Lx_cum)



ax[0][1].set_yticks([1,10,100])
ax[0][1].set_yscale('log')

ax[1][1].set_yticks([1,10,100])
ax[1][1].set_yscale('log')
```
