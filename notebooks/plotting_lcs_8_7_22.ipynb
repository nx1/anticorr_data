```python
from glob import glob
import matplotlib.pyplot as plt
import pandas as pd
```

```python
class XRTLC:
    def __init__(self, path):
        self.path    = path
        self.df      = pd.read_csv(path, dtype={'obsID':str})
        self.df_good = self.df[self.df['BAD'] != True]
        self.df_bad  = self.df[self.df['BAD'] == True]
        self.df_UL   = self.df[self.df['UL'] == True]
        self.get_lc_type()

    def get_lc_type(self):
        if 'curve_nosys' in self.path:
            self.lc_type = 'curve_nosys'
        elif 'hardrat' in self.path:
            self.lc_type = 'hardrat'

        
    
    def plot_lc(self, sequential=False):
        fig, ax = plt.subplots(figsize=(20,5), facecolor='white')
        ax.set_title(csv)
        if sequential:
            if self.lc_type == 'curve_nosys':
                ax.scatter(self.df_UL.index, self.df_UL['Rate'], color='purple', marker=r'$\downarrow$', lw=1.0, label='UL')
                ax.errorbar(self.df_bad.index, self.df_bad['Rate'], yerr=(self.df_bad['Rateneg'].abs(), self.df_bad['Ratepos']), ls='none', color='red', lw=1.0, capsize=1.0, label='BAD')
                ax.errorbar(self.df_good.index, self.df_good['Rate'], yerr=(self.df_good['Rateneg'].abs(), self.df_good['Ratepos']), ls='none', color='green', lw=1.0, capsize=1.0, label='GOOD')
            else:
                ax.scatter(self.df_UL.index, self.df_UL['Rate'], color='purple', marker=r'$\downarrow$', lw=1.0, label='UL')
                ax.errorbar(self.df_bad.index, self.df_bad['Rate'], yerr=self.df_bad['Error'], ls='none', color='red', lw=1.0, capsize=1.0, label='BAD')
                ax.errorbar(self.df_good.index, self.df_good['Rate'], yerr=self.df_good['Error'], ls='none', color='green', lw=1.0, capsize=1.0, label='GOOD')
                
        else:
            if self.lc_type == 'curve_nosys':
                ax.scatter(self.df_UL['MJD'], self.df_UL['Rate'], color='purple', marker=r'$\downarrow$', lw=1.0, label='UL')
                ax.errorbar(self.df_bad['MJD'], self.df_bad['Rate'], yerr=(self.df_bad['Rateneg'].abs(), self.df_bad['Ratepos']), ls='none', color='red', lw=1.0, capsize=1.0, label='BAD')
                ax.errorbar(self.df_good['MJD'], self.df_good['Rate'], yerr=(self.df_good['Rateneg'].abs(), self.df_good['Ratepos']), ls='none', color='green', lw=1.0, capsize=1.0, label='GOOD')
            else:
                ax.scatter(self.df_UL['MJD'], self.df_UL['Rate'], color='purple', marker=r'$\downarrow$', lw=1.0, label='UL')
                ax.errorbar(self.df_bad['MJD'], self.df_bad['Rate'], yerr=self.df_bad['Error'], ls='none', color='red', lw=1.0, capsize=1.0, label='BAD')
                ax.errorbar(self.df_good['MJD'], self.df_good['Rate'], yerr=self.df_good['Error'], ls='none', color='green', lw=1.0, capsize=1.0, label='GOOD')
                
        ax.legend()
        plt.show()
        
    def describe(self):
        res = {}
        res['path']   = self.path
        res['n_obs_df'] = len(self.df['obsID'].unique())
        res['n_obs_df_good'] = len(self.df_good['obsID'].unique())
        res['n_obs_df_bad'] = len(self.df_bad['obsID'].unique())
        res['n_obs_df_UL'] = len(self.df_UL['obsID'].unique())
        
        print(res)
        return res
```
